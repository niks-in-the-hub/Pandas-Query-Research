```python
import pandas as pd
```

```python
import numpy as np
```

```python
df=pd.DataFrame(np.random.randint(0,100,size=(100,4)),columns=list('ABCD'))  #creating dataset with random integers using numpy function
```

```python
df.head(50)  #select values for 0 to 49. i.e, 50 values
             #if tail(50), selects values for 50 to 99
```

```python
df.query("index>A") #Checks condition and displays only those corresponding A,B,C,D values where the value of index is greater
                    #than the value of A
```

```python
df.query("index<A & index>B")  #Multiple conditions can be performed under a single query()
                               #Displays results only if it accepts the condition where index<A and index>B is true.
                               #Doesn't work only if either one of them is true.
```

```python
df.query("A>30")  #checking for all values in column A and return if greater than 30
```

```python
df.query("A>30 & B<40")  #using two conditions in a single query
```

```python
df.query("A>30 & B<40").shape  #.shape gives the size of (rows,columns) of the result of a condition.
```

```python
df.query("A>30 & B<40").count
```

```python
df.iloc[60:66,0:3]  #Specify and only request for a chunk of the dataset
                    #It always displays values from "specified_start" to "specified_end-1"
```

```python
df = df.drop(df[df.A==0].index)
```

```python
df
```

```python
df.query("A in [37, 57, 98]")  #query if A has one of the specified set of values.
                               #displays all rows that has one of them
```

```python
df.query("A==37 | A==57 | A==98")  #the above can be re-written by giving conditional operators.
```

```python
df.query("A < B and B < C")  #checks condition of < first.
                             #then applies the 'and' condition.
```

```python
df.iloc[44:60,0:3]  #select a subsection
```

```python
df.iloc[44:60,0:3].query("A in [50,74]")  #method chaining query and iloc   (where A is either 50 or 74)
```

```python
df.iloc[44:60,0:3].query("index in [57,60]")   #writing query and iloc using index   (where row number is 57 or 60)
```

```python
df.iloc[24:59].query("index>50 & index<57")  #select the specified chunk of the dataset and apply a condition on it.
```

```python
df.iloc[44:60,0:1].query("A>=0")
```

```python
df.iloc[44:59]  #performing a sample iloc operation
```

```python
df.query("index>44 & index<59")  #replicating the iloc functionality as above 
```

```python
#based on above results, there appear to be subtle differences in the way iloc and query modules interpret dataframe indexing.
#Adapting the query accordingly.


df.query("index>43 & index<59")
```

```python
len(df.iloc[44:60,0:1].query("A>=0"))/len(df) *100    #percentage of data meeting this certain condition
```

```python
df.iloc[44:60,0:1].query("A>=0").shape
```

```python
df.filter(items=['A', 'C'])  #display values in columns A and C only
```

```python
len(df[df.A > 85])  #number of values that meet the condition A>85
```

```python
df.loc[df['A']>10, ['A','C']]  #Select rows meeting logical condition, and only the specific columns
```

```python
df.filter(["A","C"]) .query("A>10")                                  # df.loc[df['A']>10, ['A','C']]

#Some conventional implementation looks like      df.loc[df['A'] > 10, ['A','C']]
#in such implementations are less intuitive, and semantically unclear. A cleaner way is to chain the 'query' and 'filter'
#modules.
```
